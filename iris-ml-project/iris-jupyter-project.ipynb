```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(pd)
```

```python
iris_data = pd.read_csv("iris.csv")
```

```python
iris_data.head(10)
```

```python
print(iris_data.columns)
```

```python
iris_data['species'].value_counts()
```

```python
iris_data.isna().sum().sort_values(ascending=False)
```

```python
iris_data.isnull().sum().sort_values(ascending=False)
```

```python
# primeiro plot
plt.plot(iris_data['species'])
plt.xlabel("Qntde. de linhas")
plt.show()
```

```python
# segundo plot
plt.hist(iris_data["species"],color="green")   # to plot a histogram 
plt.show()
```

```python
print(iris_data.describe())
```

```python
sns.set_style('whitegrid');
sns.FacetGrid(iris_data,hue='species', height=5).map(plt.scatter,
'sepal_length', 'sepal_width').add_legend();
plt.show();
```

```python
sns.boxplot(x='species',y='petal_length',data=iris_data)
plt.grid()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris_data,hue='species',height=1.5)
plt.show()
```
